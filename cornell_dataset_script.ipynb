```python
lines = open('movie_lines.txt', encoding='utf-8', errors='ignore').read().split('\n')
conv_lines = open('movie_conversations.txt', encoding='utf-8', errors='ignore').read().split('\n')
```

```python
conv_lines[:10]
```

```python
WILL_HUNTING_MOVIE_ID = ' m373 '

conversations = []

for conversation in conv_lines:
    try:
        if conversation.split('+++$+++')[2] == WILL_HUNTING_MOVIE_ID :
            conversations.append(conversation)
    except:
        pass
```

```python
conversations
```

```python
from collections import defaultdict

movie_lines = defaultdict(lambda: "")

for line in lines:
    try:
        line_elements = line.split('+++$+++')
        line_id, character_id, movie_id, text = line_elements[0], line_elements[1], line_elements[2], line_elements[4] 
        if movie_id == WILL_HUNTING_MOVIE_ID:
            line_id = line_id.replace(' ', '')
            movie_lines[line_id] = (character_id.replace(' ', ''), text)
    except:
        pass
```

```python
SEAN_CHARACTER_ID = 'u5615 '
WILL_CHARACTER_ID = 'u5621 ' 

utterances = defaultdict(lambda: [])
conversation_index = 0

for conversation in conversations:
    elements = conversation.split('+++$+++')
    first_character_id, second_character_id = elements[0], elements[1]
    
    if ((first_character_id == SEAN_CHARACTER_ID or first_character_id == WILL_CHARACTER_ID) 
        or (second_character_id == SEAN_CHARACTER_ID or second_character_id == WILL_CHARACTER_ID)):
        
        conversation_index += 1
        
        lines_id = elements[3]
        lines_id = lines_id.replace('[', '')
        lines_id = lines_id.replace(']', '')
        lines_id = lines_id.replace("'", '')
        lines_id = lines_id.replace(" ", '')
        lines_id = lines_id.split(',')
        
        for line in lines_id:
            line_character = 'WILL' if movie_lines[line][0] == 'u5621' else 'SEAN'
            line_text = movie_lines[line][1]
            utterances[conversation_index].append([line_character,line_text])

for conversation in utterances.values():
    print(conversation)
    print('\n\n')
    
```

```python
with open('conversations.txt', 'w') as f:
    for conversation in utterances.values():
        to_write = ''
        for single_utterance in conversation:
            to_write += single_utterance[0] + ' + ' + single_utterance[1] + ' + '
        f.write(to_write)
        f.write('\n')
```
